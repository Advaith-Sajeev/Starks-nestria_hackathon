```python
import pandas as pd

# Replace 'file1.csv' and 'file2.csv' with your actual file names
file1_path = 'historic_performance_partial2.csv'
file2_path = 'historic_performance_partial.csv'

# Read the CSV files into pandas DataFrames
df1 = pd.read_csv(file1_path)
df2 = pd.read_csv(file2_path)

# Combine the DataFrames using the concat function
combined_df = pd.concat([df1, df2], ignore_index=True)

# Write the combined DataFrame to a new CSV file
combined_df.to_csv('combined_file2.csv', index=False)

print("Files combined successfully and saved as 'combined_file.csv'")
```
